```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Suppose this is your full dataset with 1470 rows and 3 features
X_all = np.random.rand(1470, 3)  # (1470 samples, 3 features)
y = np.random.randint(0, 2, size=1470)  # (1470 samples, binary labels)

# Scaling the features
scaler = MinMaxScaler()
X = scaler.fit_transform(X_all)  # This will have shape (1470, 3)

# Now split correctly
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Show the shapes
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
display (y_pred)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print("Accuracy {} %".format( 100 * accuracy_score(y_pred, y_test)))
```

```python
cm = confusion_matrix(y_pred, y_test)
print (cm)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
display (y_pred)
```

```python
print("Accuracy {} %".format( 100 * accuracy_score(y_pred, y_test)))
print ('\n Confusion Matrix')
cm = confusion_matrix(y_pred, y_test)
print (cm)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from keras.models import Sequential
from keras.layers import Dense

# Assuming X_train has shape (num_samples, 3)
model = Sequential()
model.add(Dense(64, input_shape=(3,), activation='relu'))  # Input layer with 3 features
model.add(Dense(32, activation='relu'))                    # Hidden layer
model.add(Dense(1, activation='linear'))                   # Output layer

model.compile(optimizer='adam', loss='mean_squared_error')

# Fit the model
epochs_hist = model.fit(X_train, y_train, epochs=100, batch_size=50)
```

```python
y_pred = model.predict(X_train)
y_pred = (y_pred > 0.5)
display (y_pred)
```
